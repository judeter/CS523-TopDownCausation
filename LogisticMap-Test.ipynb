```python
import numpy as np
import matplotlib.pyplot as plt
from LogisticGrowthModel import *
```

```python
N_transient  = 1000
N_steadyState= 3000 
N = 500 # Number of coupled logistic growth maps
r = np.random.uniform(low=0,high=4.0, size=(N,))
K = np.ones(shape=(N,))*100
growthModels = generateGrowthModels(r,K)
X = np.ones(shape=(N,1))
```

```python
X = np.zeros(shape=(N,N_steadyState))
X[:,0]=1.0
for n in range(N_transient):
    X[:,0]= globalLogisticGrowth(X[:,0],growthModels,0)
for i in range(N_steadyState-1):
    X[:,i+1]=globalLogisticGrowth(X[:,i],growthModels,0)
```

```python
for inx in range(len(r)):
    plt.plot(np.ones(X[inx,:].shape)*r[inx],X[inx,:],'o',markerfacecolor='none', markeredgecolor='k', alpha=0.01, 
             markersize=1)
plt.xlabel('Reproduction Rate', fontsize=12);
plt.ylabel('Population',fontsize=12);
```
